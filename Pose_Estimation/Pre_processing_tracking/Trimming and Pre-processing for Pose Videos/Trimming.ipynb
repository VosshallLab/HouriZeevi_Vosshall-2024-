```python
import os
import csv
import shutil
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import glob
#input_directory = "H:\MonaVideoRecordings"
#output_directory = "H:\Trimmed_encoded"
def trim_videos(input_tsv, input_directory, output_directory):
    with open(input_tsv, 'r') as tsv_file:
        reader = csv.reader(tsv_file, delimiter='\t')
        rows = list(reader)

    starting_times = []
    stopping_times = []
    video_paths = []

    for row in rows:
        if len(row) > 0:
            if row[-1] == 'START':
                starting_times.append(float(row[0]))
                video_path = row[1].replace('/Volumes/LaCie/', "{}\\".format(input_directory))  # Remove "/Volumes/" from video path
                video_paths.append(video_path)
            elif row[-1] == 'STOP':
                stopping_times.append(float(row[0]))

    for video_path, start_time, stop_time in zip(video_paths, starting_times, stopping_times):
        video_file = os.path.join(input_directory, video_path)
        video_file = os.path.normpath(video_file)  # Normalize the file path
        output_file = os.path.join(output_directory, f"{input_tsv}_{os.path.basename(video_path)}_{start_time}-{stop_time}.mp4")

        ffmpeg_extract_subclip(video_file, start_time, stop_time, targetname=output_file)

    print("Video trimming completed.")

# Get a list of TSV files in the current directory
tsv_files = glob.glob("*.tsv")

# Iterate over each TSV file and trim videos
for tsv_file in tsv_files:
    input_tsv = tsv_file  # Update with your desired output directory

    trim_videos(input_tsv, input_directory, output_directory)



```
