```python
import os
import csv
import shutil
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import glob

def trim_videos(input_tsv, input_directory, output_directory):
    tsv_file_name = os.path.splitext(os.path.basename(input_tsv))[0]  # Get the name of the TSV file without extension
    
    with open(input_tsv, 'r') as tsv_file:
        reader = csv.reader(tsv_file, delimiter='\t')
        rows = list(reader)

    starting_times = []
    stopping_times = []
    video_paths = []

    for row in rows:
        if len(row) > 0:
            if row[-1] == 'START':
                starting_times.append(float(row[0]))
                video_path = row[1].replace('/Volumes/LaCie/', "{}\\".format(input_directory))  # Remove "/Volumes/" from video path
                video_paths.append(video_path)
            elif row[-1] == 'STOP':
                stopping_times.append(float(row[0]))

    for video_index, (video_path, start_time, stop_time) in enumerate(zip(video_paths, starting_times, stopping_times)):
        video_file = os.path.join(input_directory, video_path)
        video_file = os.path.normpath(video_file)  # Normalize the file path
        output_file = f"{tsv_file_name}_sub{video_index}.mp4"  # Output file name with TSV file name and index

        # Calculate the trimmed start and stop times
        trim_start_time = max(0, start_time - 3)
        if stop_time < start_time + 5:
            trim_stop_time = stop_time
        else:
            trim_stop_time = start_time + 5

        ffmpeg_extract_subclip(video_file, trim_start_time, trim_stop_time, targetname=os.path.join(output_directory, output_file))

    print("Video trimming completed.")

# Specify the directory where TSV files are located
tsv_directory = r"H:\GenitalContactPlusMatingAnnotations"

# Specify input and output directories for trimmed videos
input_directory = r"H:\MonaVideoRecordings"
output_directory = r"H:\Novel_Trimming_Videos"

# Get a list of TSV files in the specified directory
tsv_files = glob.glob(os.path.join(tsv_directory, "*.tsv"))

# Iterate over each TSV file and trim videos
for tsv_file in tsv_files:
    trim_videos(tsv_file, input_directory, output_directory)

```
